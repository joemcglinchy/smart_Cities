```python
# check if geometries are equal and create data frame that has unique geometries
import shapely

gdf2 = gdf.copy()

a = shapely.geometry.Polygon(gdf['geometry'][0])
b = shapely.geometry.Polygon(gdf['geometry'][0])

a.almost_equals(b)

ls = []
unique_geoms = [gdf['geometry'][0]]
for i,row in gdf2.iterrows():
    
    if row['geometry'] not in unique_geoms:
        ls.append(row)
        unique_geoms.append(row['geometry'])
    else:
        continue
        
        
    
    
```

```python
filt = gpd.geodataframe.GeoDataFrame(ls)
filt.reset_index(drop=True)
filt.head()
filt.plot()
```
